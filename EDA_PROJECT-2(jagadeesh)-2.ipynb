# 3.Provide a summary of the cleaned dataset along with some preliminary analysis and contrast it with the uncleaned dataset wherever modifications are made.

```python
# Removing NaN values from Dataframe i.e cleaning the data
# First we are checking if there are any rows with NaN
nan_df=df[df.isna().any(axis=1)]
nan_df.head(10)
```

```python
# here we are removing NaN values from the Dataframe
nan_df=df[df.isna().any(axis=1)]
nan_df.head()
df=df.dropna(how='all')
df.head()
```

```python
#an error appeared when solving ahead stating there was 'or' in the order date to remove this error
#We are first checking which rows have 'or' in them

temp_df=df[df['Order Date'].str[0:2]== 'Or']
temp_df
```

```python
#removing the rows with 'or' in them
df=df[df['Order Date'].str[0:2]!= 'Or']
df.head()
```

```python
# converting Quantity Ordered to int values
df['Quantity Ordered']=pd.to_numeric(df['Quantity Ordered'])
# converting Price Each to float values
df['Price Each']=pd.to_numeric(df['Price Each'])
```

```python
# getting info of the data frame.
df.info()
```

```python
# getting data types of data frames
df.dtypes
```

```python
# adding month column to the data frame by using Order Date column
df['Month']=df['Order Date'].str[0:2]
df['Month']=df['Month'].astype('int32')
df.head()
```

```python
# adding sales column to the data frame by using Quantity Ordered * Price Each
df['sales']=df['Quantity Ordered']* df['Price Each']
df.head()
```

```python
# adding city column by extracting Purchase Address column
df['City']=df['Purchase Address'].apply(lambda x:x.split(',')[1])
df.head()
```

```python
# here we are extracting the city, state and neglecting the pincode from the Purchase Address.  
def get_city(address):return address.split(',')[1]

def get_state(address):return address.split(',')[2].split(' ')[1]

df['City']=df['Purchase Address'].apply(lambda x:get_city(x) + ' ' + get_state(x))
df.head()
```

# 4.What is the distribution of sales among different cities in different months?

# a.Which city had the highest sales overall?

```python
# here we are finding the highest sales of city.using groupby function.
results=df.groupby('City').sum()
results
```

```python
# plotting results
cities= [city for city,df in df.groupby('City')]
# by using bar plot we are plotting.
plt.bar(cities,results['sales'])
plt.xticks(cities,rotation='vertical',size=9)
plt.ylabel('sales in usd($)')
plt.xlabel('City name')
plt.show()
```

```python
# san francisco ca has the highest sales overall.
```

# b.Which Month had the highest amount of sales overall?

```python
# here we are finding the highest amount of sales in month
results=df.groupby('Month').sum()
results
```
